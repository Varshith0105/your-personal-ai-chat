```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, pipeline

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
    llm_int8_enable_fp32_cpu_offload=True
)

model_id = "meta-llama/Meta-Llama-3-8B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_id)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    quantization_config=bnb_config,
    torch_dtype=torch.float16
)
```

## ***Creating Conservational Memory***

```python
chat_history=[]
```

```python
def build_prompt(user_input):
  history = "\n".join(chat_history[-6:])

  prompt = f"""
You are a helpful AI assistant.

Conversation history:
{history}

user: {user_input}
Assistant:
  """
  return prompt
```

```python
def chat(user_input):
  prompt = build_prompt(user_input)

  response = llama_pipeline(
      prompt,
      max_new_tokens=300,
      do_sample=True,
      temperature=0.7
  )[0]["generated_text"]

  answer = response[len(prompt):].strip()

  chat_history.append(f"user: {user_input}")
  chat_history.append(f"Assistant: {answer}")

  return answer
```

```python
print(chat("Hi,Who are you"))
```

*Building the LangGraphs*

```python
from langgraph.graph import StateGraph
```

```python
#Defining the Agent state
class AgentState(dict):
  input: str
  output: str
```

```python
#Creating the nodes with the history backup
def reasoning_node(state):
  prompt = f"""
  You are an intelligent AI assistance.
  Think step-by-step before answering.

User question:
{state['input']}

Final answer:
"""
  result = llama_pipeline(
      prompt,
      max_new_tokens=300,
      do_sample=True,
      temperature=0.7
  )[0]["generated_text"]

  answer = result.split("Final answer")[-1].strip()
  state['output'] = answer


  return {"output": answer}
```



*> Here comes with the main part of the Building the Langgraphs*



```python
graph = StateGraph(AgentState)

graph.add_node("reason",reasoning_node)

graph.set_entry_point("reason")
graph.set_finish_point("reason")

agent = graph.compile()
```

```python
response = agent.invoke({"input": "Explain what are the Transformers in a simple English"})
print(response["output"])
```

```python
import requests
```

```python
Weather_API_KEY = "09b228f646e02a94712901f6431aef1c"

def Weather(city):
  url = 'https://api.openweathermap.org/data/2.5/weather'
  params = {
      "q" : city,
      "appid" : Weather_API_KEY,
      "units" : "metric"
      "weather_description": "Clear conditions throughout the day.",
  }


  return requests.get(url, params=params).json()
```

```python
News_API_KEY = "5a3812b50ca64758a4b42ceeb05d3ec6"

def news(topic):
  url='https://newsapi.org/v2/everything'
  params={
      "q": topic,
      "apikey": News_API_KEY,
      "Pagesize": 5
  }

  return requests.get(url, params=params).json()
```

```python
def tool_router(state):
  q = state["input"].lower()

  if "weather" in q:
    return {"next_step": "Weather"}
  elif "news" in q or "headlines" in q:
    return {"next_step": "news"}
  else:
    return {"next_step": "llm"}
```

***Creating the execution tool nodes***

```python
def Weather_node(state):
  city = state["input"].split()[-1]
  data = Weather(city)
  return {"tool_data": str(data)}

def news_node(state):
  topic = state["input"].split()[-1]
  data = news(topic)
  return {"tool_data": str(data)}

def llm_node(state):
  history = state.get('history', '')
  prompt = f"""
  You are a helpful AI assistant.

  Conversation history:
  {history}

  User question: {state['input']}
  Answer:
  """
  result = llama_pipeline(
      prompt,
      max_new_tokens=300,
      do_sample=True,
      temperature=0.7
  )[0]["generated_text"]
  answer = result[len(prompt):].strip()
  return {"output": answer}
```

```python
def answer_node(state):
  history = state.get('history', '')
  prompt = f"""

        You are a helpful AI assistant

  Conversation history:
  {history}

  User Question:
  {state["input"]}

  Tool_data:
  {state["tool_data"]}

  Answer clearly using the tool data if available.
  """

  response=llama_pipeline(
      prompt,
      max_new_tokens=300,
      temperature=0.7,
      do_sample=True
  )[0]["generated_text"]

  answer = response[len(prompt):].strip()
  return {"output": answer}
```

 After creating the Executing Tools, The next Step is used to Create a LangGraphs with the created Tools


```python
from langgraph.graph import StateGraph
```

```python
from typing import TypedDict
from langgraph.graph import StateGraph

class AgentState(TypedDict):
  input: str
  tool_data: str
  output: str
  history: str
  next_step: str

graph = StateGraph(AgentState)

graph.add_node("router", tool_router)
graph.add_node("Weather", Weather_node)
graph.add_node("news", news_node)
graph.add_node("llm", llm_node)
graph.add_node("answer", answer_node)

graph.set_entry_point("router")

graph.add_conditional_edges(
    "router",
    lambda state: state['next_step'],
    {
        "Weather": "Weather",
        "news": "news",
        "llm": "llm"
    }
)

graph.add_edge("Weather", "answer")
graph.add_edge("news", "answer")
graph.add_edge("llm", "answer")

graph.set_finish_point("answer")
agent = graph.compile()
```

```python
print(agent.invoke({"input": "Give me the Planks constant number"})["output"])
```

```python
def agent_chat_interface(user_input, chat_history):
    history_str = ""
    for human_msg, ai_msg in chat_history:
        history_str += f"user: {human_msg}\nAssistant: {ai_msg}\n"

    response = agent.invoke({"input": user_input, "history": history_str})
    return response["output"]
```

```python
import gradio as gr

chat_interface = gr.ChatInterface(
    fn=agent_chat_interface,
    chatbot=gr.Chatbot(height=400),
    textbox=gr.Textbox(placeholder="Ask me a question...", container=False, scale=7),
    title="LangGraph Agent Chatbot",
    description="Ask me questions about the weather, news, or general topics!",
    theme="soft",
    examples=["What is the weather in London?", "Give me today's headlines", "Explain quantum physics"]
)

chat_interface.launch(share=True)
```
